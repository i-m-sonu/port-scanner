```python
import socket
from IPy import IP
import threading

# Shared lists with thread synchronization
ports = []
banners = []
lock = threading.Lock()

def port_scanner(target, port):
    try:
        # Create and configure socket
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(2)
        try:
            # Resolve target to IP
            target_ip = IP(target)
        except ValueError:
            target_ip = socket.gethostbyname(target)

        # Attempt to connect to the port
        s.connect((str(target_ip), port))
        try:
            # Get banner if available
            banner_name = banner(s).decode('utf-8').strip()
            with lock:  # Ensure thread-safe operations
                ports.append(port)
                banners.append(banner_name)
        except:
            pass
        s.close()
    except:
        pass

def banner(s):
    # Attempt to retrieve the banner
    return s.recv(1024)

def main():
    target = input("Enter Target IP address, localhost, or www.domainname.com: ")
    portsc = int(input("Enter Number of ports to be scanned: "))

    # Launch threads for port scanning
    threads = []
    for port in range(1, portsc + 1):
        thread = threading.Thread(target=port_scanner, args=(target, port))
        threads.append(thread)
        thread.start()

    # Wait for all threads to complete
    for thread in threads:
        thread.join()

    # Check for vulnerabilities in collected banners
    try:
        with open("vulnarable_banners.txt", "r") as file:
            vulnerable_banners = file.read().splitlines()
            for i in range(len(banners)):
                if banners[i] in vulnerable_banners:
                    print(f"[!] Vulnerability found: {banners[i]} at port {ports[i]}")
    except FileNotFoundError:
        print("[!] Vulnerable banners file not found!")

if __name__ == "__main__":
    main()
```

```python
import socket
from IPy import IP
import threading

# Shared lists with thread synchronization
ports = []
banners = []
lock = threading.Lock()

# List of known vulnerable banners
vulnerable_banners = [
    "SSH-2.0-OpenSSH_7.4",
    "Apache HTTPD 2.4.1",
    "OpenSSL/1.0.2g",
    "nginx/1.14.0",
    "FTP/1.0",
    "MySQL 5.7.20"
]

def port_scanner(target, port, total_ports):
    try:
        # Create and configure socket
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(2)
        try:
            # Resolve target to IP
            target_ip = IP(target)
        except ValueError:
            target_ip = socket.gethostbyname(target)

        # Display progress for the current port
        print(f"[+] Scanning port {port}/{total_ports}...", end="\r")

        # Attempt to connect to the port
        s.connect((str(target_ip), port))
        try:
            # Get banner if available
            banner_name = banner(s).decode('utf-8').strip()
            with lock:  # Ensure thread-safe operations
                ports.append(port)
                banners.append(banner_name)
                print(f"[+] Port {port} open: {banner_name}")
        except:
            print(f"[+] Port {port} open, but no banner found.")
        s.close()
    except:
        print(f"[-] Port {port} closed.")

def banner(s):
    # Attempt to retrieve the banner
    return s.recv(1024)

def main():
    target = input("Enter Target IP address, localhost, or www.domainname.com: ")
    portsc = int(input("Enter Number of ports to be scanned: "))

    print(f"\n[+] Starting scan for {target} on {portsc} ports...\n")

    # Launch threads for port scanning
    threads = []
    for port in range(1, portsc + 1):
        thread = threading.Thread(target=port_scanner, args=(target, port, portsc))
        threads.append(thread)
        thread.start()

    # Wait for all threads to complete
    for thread in threads:
        thread.join()

    # Check for vulnerabilities in collected banners
    print("\nScanning completed. Checking for vulnerabilities...\n")
    for i in range(len(banners)):
        if banners[i] in vulnerable_banners:
            print(f"[!] Vulnerability found: {banners[i]} at port {ports[i]}")

if __name__ == "__main__":
    main()
```

```python
import socket
from IPy import IP
import threading

# Shared resources for thread-safe operations
open_ports = []
service_banners = []
lock = threading.Lock()

# List of known vulnerable services
vulnerable_services = [
    "SSH-2.0-OpenSSH_7.4",
    "Apache HTTPD 2.4.1",
    "nginx/1.14.0",
    "ProFTPD 1.3.5",
    "MySQL 5.7.20"
]

def scan_port(target, port):
    try:
        # Create a socket
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(1)

        # Resolve the target IP address
        try:
            target_ip = IP(target)
        except ValueError:
            target_ip = socket.gethostbyname(target)

        # Try to connect to the port
        result = s.connect_ex((str(target_ip), port))
        if result == 0:  # Port is open
            try:
                # Attempt to retrieve the service banner
                s.send(b'Hello\r\n')  # Send dummy data
                banner = s.recv(1024).decode('utf-8').strip()
            except:
                banner = "No banner available"

            # Store the results in shared resources
            with lock:
                open_ports.append(port)
                service_banners.append(banner)
                print(f"[+] Port {port} is open: {banner}")
        s.close()
    except Exception as e:
        pass  # Ignore errors for now

def scan_target(target, port_range):
    print(f"[+] Scanning target {target} for ports 1 to {port_range}...\n")

    threads = []
    for port in range(1, port_range + 1):
        # Launch threads for port scanning
        thread = threading.Thread(target=scan_port, args=(target, port))
        threads.append(thread)
        thread.start()

    # Wait for all threads to complete
    for thread in threads:
        thread.join()

    print("\n[+] Scan completed.")
    print("[+] Open Ports and Services Detected:")
    for i in range(len(open_ports)):
        print(f"    Port {open_ports[i]}: {service_banners[i]}")

    # Check for vulnerabilities
    print("\n[+] Checking for vulnerabilities...")
    for i in range(len(service_banners)):
        if service_banners[i] in vulnerable_services:
            print(f"[!] Vulnerable service detected: {service_banners[i]} on port {open_ports[i]}")
    print("[+] Vulnerability check completed.")

def main():
    target = input("Enter the target IP or domain (e.g., 192.168.1.1 or www.example.com): ")
    port_range = int(input("Enter the number of ports to scan (e.g., 100): "))
    scan_target(target, port_range)

if __name__ == "__main__":
    main()
```
